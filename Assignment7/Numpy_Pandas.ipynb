```python
# Exercise 3: (Score : 2) 
# Create a Pandas series with the following data: {'apples': 3, 'bananas': 2, 'oranges': 1}. Then, add a new item to the series 
# with the key 'pears' and the value 4.

import pandas as pd

data = {'apples': 3, 'bananas': 2, 'oranges': 1}
df = pd.Series(data)
print(df,'\n')

df['pears']=4
print("After Updating") 
df
```

```python
# Exercise 4: (Score : 2) 
# Create a dataframe with the following columns: name, age, and gender. The dataframe should have 10 rows of data. 


dt = {   
    "Name": ['Adarsh','Arjun','Aswathi','Ajay','Diya','Rahul','Thejus','Harsha','Sharuk','Nandu'],
    "Age": [23,32,19,25,20,23,34,18,25,30],
    "Gender": ['M','M','F','M','F','M','M','F','M','M']
}

```
